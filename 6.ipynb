```python
def draw_field(pole):
    for row in pole:
        print(' '.join(row))

def proverka_coordinates(x, y):
    if x < 0 or x >= len(pole[0]):
        return False
    if y < 0 or y >= len(pole):
        return False
    return True
def calculate_schet(x, y):
    schet = 0
    symbol = pole[y][x]
    if x > 0 and pole[y][x - 1] == symbol:
        schety += 1
    if x < len(pole[0]) - 1 and pole[y][x + 1] == symbol:
        schet += 1
    if y > 0 and pole[y - 1][x] == symbol:
        schet += 1
    if y < len(pole) - 1 and pole[y + 1][x] == symbol:
        schet += 1
    if x > 0 and y > 0 and pole[y - 1][x - 1] == symbol:
        schet += 1
    if x < len(pole[0]) - 1 and y > 0 and pole[y - 1][x + 1] == symbol:
        schet += 1
    if x > 0 and y < len(pole) - 1 and pole[y + 1][x - 1] == symbol:
        schet += 1
    if x < len(pole[0]) - 1 and y < len(pole) - 1 and pole[y + 1][x + 1] == symbol:
        schet += 1

    return schet


pole =   [['-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-'],
         ['-', '-', '-', '-', '-']]

players = ['X', 'O', '#']
schets = [0, 0, 0]

draw_field(pole)

for turn in range(len(pole[0]) * len(pole)):
    player = players[turn % len(players)]

    while True:
        x = int(input(f"Игрок {player}, введите номер столбца (1-5): "))
        y = int(input(f"Игрок {player}, введите номер строки (1-4): "))

        if not proverka_coordinates(x - 1, y - 1):
            print("Некорректные координаты! Попробуйте еще раз.")
            continue

        if pole[y - 1][x - 1] != '-':
            print("Эта клетка уже занята! Попробуйте еще раз.")
            continue

        break

    pole[y - 1][x - 1] = player
    schets[turn % len(players)] += calculate_schet(x - 1, y - 1)
    draw_field(pole)

min_schet = min(schets)
winners = [players[i] for i, schet in enumerate(schets) if schet == min_schet]

print("Игра окончена!")
print(f"Победители: {' '.join(winners)}")
print(f"Штрафные очки: {schets}")
```
