```python
#Написать калькулятор для строковых выражений вида '<число> <операция> <число>', где <число> - не отрицательное целое число меньшее 100, записанное словами, например "тридцать четыре", <арифмитическая операция> - одна из операций "плюс", "минус", "умножить". Результат выполнения операции вернуть в виде текстового представления числа. Пример calc("двадцать пять плюс тринадцать") -> "тридцать восемь"
#Оформить калькулятор в виде функции, которая принимает на вход строку и возвращает строку.
def tonumb(word):
    dict_numbers_reverse = {
    'один': 1, 'два': 2, 'три': 3, 'четыре': 4, 'пять': 5, 'шесть': 6, 'семь': 7, 'восемь': 8, 'девять': 9,
    'десять': 10, 'одиннадцать': 11, 'двенадцать': 12, 'тринадцать': 13, 'четырнадцать': 14, 'пятнадцать': 15,
    'шестнадцать': 16, 'семнадцать': 17, 'восемнадцать': 18, 'девятнадцать': 19, 'двадцать': 20, 'тридцать': 30,
    'сорок': 40, 'пятьдесят': 50, 'шестьдесят': 60, 'семьдесят': 70, 'восемьдесят': 80, 'девяносто': 90,
    'сто': 100, 'двести': 200, 'триста': 300, 'четыреста': 400, 'пятьсот': 500, 'шестьсот': 600, 'семьсот': 700,
    'восемьсот': 800, 'девятьсот': 900, 'тысяча': 1000, 'две': 2, 'три': 3, 'четыре': 4, 'пять': 5, 'шесть': 6,
    'семь': 7, 'восемь': 8, 'девять': 9, 'тысячи': 1000, 'тысяч': 1000}
    words = word.split()
    number = 0
    i = 0
    while i < len(words):
        if words[i] == 'тысяча' or words[i] == 'тысячи' or words[i] == 'тысяч':
            if i == 0 or (i > 0 and words[i - 1] != 'тысячи'):
                number *= dict_numbers_reverse[words[i]]
        elif words[i] == 'тысячи':
            pass
        elif words[i] in dict_numbers_reverse:
            number += dict_numbers_reverse[words[i]]
        else:
            print("Введенное слово не является числом")
            break
        i += 1
    return number
a1 = input("Введите число от 1 до 99 словами: ").lower()
c1=input("Введите операцию: ")
b1 = input("Введите число от 1 до 99 словами: ").lower()
a=tonumb(a1)
b=tonumb(b1)
def operation(oper):
    if oper=="сложение" or oper=="сложить" or oper=="+" or oper=="плюс":
        c=a+b
    elif oper=="вычитание" or oper=="вычесть" or oper=="-" or oper=="минус":
        c=a-b
    elif oper=="умножение" or oper=="умножить" or oper=="*":
        c=a*b
    return c
def totext(num):
    num_dict = {
        0: 'ноль', 1: 'один', 2: 'два', 3: 'три', 4: 'четыре', 5: 'пять', 6: 'шесть', 7: 'семь', 8: 'восемь', 9: 'девять',
        10: 'десять', 11: 'одиннадцать', 12: 'двенадцать', 13: 'тринадцать', 14: 'четырнадцать', 15: 'пятнадцать', 16: 'шестнадцать', 17: 'семнадцать', 18: 'восемнадцать', 19: 'девятнадцать',
        20: 'двадцать', 30: 'тридцать', 40: 'сорок', 50: 'пятьдесят', 60: 'шестьдесят', 70: 'семьдесят', 80: 'восемьдесят', 90: 'девяносто',
        100: "сто", 200:"двести", 300:"триста", 400:"четыреста", 500:"пятьсот", 600:"шестьсот", 700:"семьсот", 800:"восемьсот", 900:"девятьсот", 
        1000:"тысяча", 2000:"две тысячи", 3000:"три тысячи", 4000:"четыре тысячи", 5000:"пять тысяч", 6000:"шесть тысяч", 7000:"семь тысяч", 8000:"восемь тысяч", 9000:"девять тысяч"}
    if num in num_dict:
        num1=num_dict[num]
    if num<100 and num>20:
        des=num//10
        ed=num%10
        num1=num_dict[des*10]+" "+num_dict[ed]
    if num<1000 and num>100:
        sot=num//100
        des=(num//10)%10
        ed=num%10
        if des==0:
            num1=num_dict[sot*100]+' '+num_dict[ed]
        else:
            num1=num_dict[sot*100]+" "+num_dict[des*10]+" " +num_dict[ed]
    if num>1000:
        ts=num//1000
        sot=(num//100)%10
        des=(num//10)%10
        ed=num%10
        if sot==0 and des==0 and ed==0:
            num1=num_dict[ts*1000]
        elif sot==0 and des==0:
            num1=num_dict[ts*1000]+" "+num_dict[ed]
        elif sot==0 and ed==0:
            num1=num_dict[ts*1000]+" "+num_dict[des*10]
        elif ed==0 and des==0:
            num1=num_dict[ts*1000]+" "+num_dict[sot*100]
        elif sot==0:
            num1=num_dict[ts*1000]+" "+num_dict[des*10]+" "+num_dict[ed]
        elif des==0:
            num1=num_dict[ts*1000]+" "+num_dict[sot*100]+" "+num_dict[ed]
        elif ed==0:
            num1=num_dict[ts*1000]+" "+num_dict[sot*100]+" "+num_dict[des*10]
        elif sot!=0 and des!=0 and ed!=0:
            num1=num_dict[ts*1000]+" "+num_dict[sot*100]+" "+num_dict[des*10]+" "+num_dict[ed]
    return num1
c=operation(c1)
print("Результат:",totext(c))
    
    
    
    
    
    
```
